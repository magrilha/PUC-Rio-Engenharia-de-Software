## Operadores aritméticos e expressões

Operadores aritméticos são os símbolos que dizem ao computador qual operação matemática realizar sobre um ou mais valores.

Operadorㅤ(Operação)

Operator ㅤ(Operation)

ㅤ ● ㅤ + ㅤ(soma(addition));

ㅤ ● ㅤ - ㅤ(subtração(subtraction));

ㅤ ● ㅤ * ㅤ(multiplicação(multiplication));

ㅤ ●ㅤ **ㅤ(exponenciação(exponentiation));

ㅤ ● ㅤ / ㅤ(divisão(division)) e // (divisão inteira(integer division));

ㅤ ● ㅤ%ㅤ(resto da divisão(remainder of the division - modulo)) .

Uma expressão é a combinação de valores (como 5 ou 10), variáveis (como preco ou idade) e operadores (simbolos) que, quando processados pelo computador, resultam em um único valor.

A precedência dos operadores aritméticos:

1ª) Exponenciação (** ou ^).

2ª) Multiplicação (*), divisão (/), divisão inteira (//), resto/módulo (%).

3ª) Soma (+)e subtração (-).

Em caso de empate na precedência, será realizada a operação mais à esquerda. Mas, assim como na matemática, a utilização de parênteses pode alterar a precedência.

O "Fura-Fila": Os Parênteses ()

Existe um elemento acima de todos esses: os parênteses. Tudo o que estiver dentro de parênteses é calculado em 1º lugar absoluto, ignorando a tabela padrão.

Sem parênteses:

                  10 + 2 * 3 ** 2
                  └┬┘ └┬┘ └──┬──┘
                   │   │     1º: Exponenciação (3² = 9)
                   │   2º: Multiplicação (2 * 9 = 18)
                   3º: Soma (10 + 18)
                  Resultado = 28

```python
10 + 2 * 3 ** 2
```

Com parênteses:

                  (10 + 2) * 3 ** 2
                  └───┬───┘
                      1º: Parênteses primeiro (10 + 2 = 12)
                      2º: Exponenciação (3² = 9)
                      3º: Multiplicação (12 * 9)
                  Resultado = 108

```python
(10 + 2) * 3 ** 2
```
